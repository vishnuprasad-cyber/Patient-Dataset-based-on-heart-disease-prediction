```python
# Function to remove outliers using the IQR method except for one column
def remove_outliers_iqr(df_resampled, exclude_column):
    for col in df_resampled.columns:
        # Skip the specified column
        if col == exclude_column:
            continue
        Q1 = df_resampled[col].quantile(0.25)
        Q3 = df_resampled[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        # Filter the DataFrame to remove rows with outliers in each column
        df_resampled = df_resampled[(df_resampled[col] >= lower_bound) & (df_resampled[col] <= upper_bound)]
    return df_resampled

# Specify the column to exclude
df_resampled_cleaned = remove_outliers_iqr(df, exclude_column='HadHeartAttack')
d_resampledf=df_resampled_cleaned
```

```python
from sklearn.model_selection import KFold
#initialize kflods
kf=KFold(n_splits=5,shuffle=True,random_state=42)
#model setup
model=LogisticRegression()

#Test best fold data
best_accuracy=0
best_fold=-1

#save best fold
best_x_test=None
best_x_train=None
best_y_test=None
best_y_train=None
best_y_pred=None

for fold,(train_index,test_index) in enumerate(kf.split(x),1):
  print(f'\nFold{fold}')
  x_train,x_test=x.iloc[train_index],x.iloc[test_index]
  y_train,y_test=y.iloc[train_index],y.iloc[test_index]

  #TRAIN MODEL
  model.fit(x_train,y_train)

  y_pred=model.predict(x_test)

  acc=model.score(x_test,y_test)
  print('accuracy:',acc)

  #update best fold
  if acc > best_accuracy:
    best_accuracy=acc
    best_fold=fold
    best_x_test=x_test
    best_y_test=y_test
    best_x_train=x_train
    best_y_train=y_train
    best_y_pred=y_pred

#output for best detail
print(f'\n best fold:{best_fold} with best accuracy:{best_accuracy}')
```
